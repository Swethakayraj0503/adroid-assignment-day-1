```python
print(' Subset Selection Problem ')
print()
print('1.Problem Statement:')
print('Find all the subsets from a set of numbers whose sum is zero.')
print('Constraint: Subset size must be 5')
print('Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}')

import itertools

Set = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
SetSize = 5
ResultList = set()
subsets_of_five = itertools.combinations(Set, SetSize)

for subset in subsets_of_five:
    if sum(subset) == 0:
        ResultList.add(subset)


for r in ResultList:
    print(r)

print("\nTotal Sets: ", len(ResultList))

```

```python
print(' Subset Selection Problem ')
print()
print('1.Problem Statement:')
print('Find all the subsets from a set of numbers whose sum is zero.')
print('Constraint: Subset size must be 5')
print('Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}')
print()

import itertools
numbers = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
subsets_of_five = itertools.combinations(numbers, 5)
zero_sum_subsets = [subset for subset in subsets_of_five if sum(subset) == 0]
for subset in zero_sum_subsets:
    print(subset)
```

```python
print(' 2. Problem Statement: ')
print('Find all the subsets from a set of numbers whose sum is zero.')
print('Constraint: Subset size must be 3 to 6 only')
print('Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}')
print()

import itertools

numbers = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

subset_sizes = range(3, 7)

result_list = set()

for size in subset_sizes:
    subsets = itertools.combinations(numbers, size)
    for subset in subsets:
        if sum(subset) == 0:
            result_list.add(subset)
for r in result_list:
    print(r)

print("\nTotal Sets: ", len(result_list))
```

```python
print(' 2. Problem Statement: ')
print('Find all the subsets from a set of numbers whose sum is zero.')
print('Constraint: Subset size must be 3 to 6 only')
print('Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}')
print()

import itertools

Set = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
SetLB = 3
SetUB = 6
ResultList = set()

for size in range(SetLB, SetUB + 1):
    subsets = itertools.combinations(Set, size)
    for subset in subsets:
        if sum(subset) == 0:
            ResultList.add(tuple(sorted(subset)))
for r in ResultList:
    print(r)

print("\nTotal Sets: ", len(ResultList))
```
