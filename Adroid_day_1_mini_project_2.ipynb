```python
print('ComboFinder')
print()
print('Problem Statement:')
print('Find all the list of products whose sum-of-price is between 290 and 310.')
print('ProductList = {p1:10, p2:15, p3:20, p4:25, p5:30, p6:35, p7:50}')
print()

import itertools

ProductList = {'p1': 10, 'p2': 15, 'p3': 20, 'p4': 25, 'p5': 30, 'p6': 35, 'p7': 50,
               'p8': 40, 'p9': 55, 'p10': 60, 'p11': 65, 'p12': 75, 'p13': 70,
               'p14': 45}
LB = 290
UB = 310
ResultList = set()
product_names = list(ProductList.keys())

for size in range(2, len(ProductList)):
    for combo in itertools.combinations(product_names, size):
        combo_sum = sum(ProductList[p] for p in combo)
        if LB <= combo_sum <= UB:
            ResultList.add(tuple(sorted(combo)))
for r in ResultList:
    print(r)
print("\nTotal Sets: ", len(ResultList), "\n")
```
